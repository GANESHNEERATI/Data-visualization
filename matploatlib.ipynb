```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
#using scater plot
plt.scatter(x,y,c='r')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('2D image')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
##ploat images

plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('2D image')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g*',linewidth=2,markersize=12)
plt.subplot(2,2,2,)
plt.plot(x,y,'r--',linewidth=2,markersize=12)
plt.subplot(2,2,3)
plt.plot(x,y,'b*',linewidth=2,markersize=12)
plt.subplot(2,2,4)
plt.plot(x,y,'yo',linewidth=2,markersize=12)
```

```python
##more examples
x=np.arange(1,11)
y=3*x+5
plt.title('complex diagram')
plt.xlabel('X-lable')
plt.ylabel('Y-lable')
plt.plot(x,y,'r*-',linewidth=2,markersize=12)
```

```python
##sinwave draw
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('Sinewave diagram')
plt.plot(x,y,'b*')
```

```python
##sin and cos waves diagram
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
##creating subploat
plt.subplot(2,1,1)
plt.plot(x,y_sin,'g*',linewidth=2,markersize=12)
plt.title('sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'b*',linewidth=2,markersize=12)
plt.title('cosine')
```

```python
##bar graphs
x=[8,9,3]
y=[3,5,7]

x2=[5,7,8]
y2=[9,7,3]

plt.bar(x,y,color='y')
plt.bar(x2,y2,color='g')

plt.title('Bar graphs')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

```

```python
##histograms

a=np.array([5,98,21,45,32,45,76,52,85,20])
plt.hist(a,bins=10)
plt.title('histograms')
```
